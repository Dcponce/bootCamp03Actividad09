# Machine Learning: Bank Dataset

## Introducción

### Usando Machine Learning para Predecir la Suscripción a Depósitos a Plazo Fijo de Clientes Bancarios con Scikit-Learn

El dataset se basa en las campañas de marketing directo de un banco portugues. Estas campañas de marketing se realizaron mediante llamadas telefónicas. Se requería más de un contacto con un cliente para hacer toda la gestión de venta del producto (depósito a plazo fijo). El objetivo de la clasificación es predecir si un cliente suscribirá el depósito a plazo fijo del banco (sí/no). Es decir nuestra variable Target sera la columna "Y", la cual nos indica si el cliente tomo o no el producto



## Dataset

### Variables de Entrada:

1 - age (numeric)

2 - job : tipo de trabajo (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')

3 - marital : Estado Civil (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)

4 - education (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')

5 - default: Ha caido en impago? (categorical: 'no','yes','unknown')

6 - housing: Tiene un crédito hipotecario? (categorical: 'no','yes','unknown')

7 - loan: Tiene un crédito personal? (categorical: 'no','yes','unknown')

### Campos asociados a la campaña:

8 - contact: Tipo de comunicación (categorical: 'cellular','telephone') 

9 - month: Mes de la ultima llamada (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')

10 - day_of_week: dia de la ultima llamada (categorical: 'mon','tue','wed','thu','fri')

11 - duration: Duración en segundos de la ultima llamada (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

### Other attributes:

12 - campaign: cantidad de llamadas realizadas a cada cliente (numeric, includes last contact)

13 - pdays: cantidad de dias que transcurrieron previo a la ultima llamada (numeric; 999 means client was not previously contacted)

14 - previous: Cantidad de llamadas realizadas antes de esta campaña (numeric)

15 - poutcome: Resultado de la campaña de llamadas anterior (categorical: 'failure','nonexistent','success')

### Social and economic context attributes:

16 - emp.var.rate: Tasa de variación de empleo - quarterly indicator (numeric)

17 - cons.price.idx: Indice de precios al consumidor - monthly indicator (numeric) 

18 - cons.conf.idx: Indice de confianza - monthly indicator (numeric) 

19 - euribor3m: tasa Euribor a 3 meses - daily indicator (numeric)

20 - nr.employed: Número de empleados - quarterly indicator (numeric)

### Variable de Salida (target):

21 - y - Ha el cliente tomado el producto bancario ofrecido? (binary: 'yes','no')

Fuente:https://archive.ics.uci.edu/ml/machine-learning-databases/00222/ 

Dataset 40,000+ registros.



### Otros conceptos:

**El Euribor** (Euro Interbank Offered Rate) es una tasa de interés de referencia que refleja el promedio de las tasas a las cuales un grupo de bancos europeos se prestan dinero entre sí en el mercado interbancario.

**Un índice** es una medida estadística que representa el cambio en el valor o nivel de un conjunto de variables. Los índices se utilizan comúnmente en economía y finanzas para medir el desempeño de mercados, precios, niveles de producción, confianza de los consumidores, y más. Los índices permiten comparar cambios a lo largo del tiempo de manera estandarizada.

## One-Hot Encoding

El one-hot encoding es una técnica utilizada para convertir variables categóricas en una representación numérica que pueda ser utilizada en algoritmos de machine learning. Aquí te explico cómo funciona:

Definición de Categorías: Primero, se identifican todas las categorías posibles de la variable categórica. Por ejemplo, si tienes una variable "Color" con valores posibles como "Rojo", "Verde" y "Azul", estas son las categorías.

Creación de Columnas: Se crea una columna binaria (0 o 1) para cada categoría. En el caso de la variable "Color", se crearían tres nuevas columnas: "Color_Rojo", "Color_Verde" y "Color_Azul".

Asignación de Valores: Para cada fila del dataset, se asigna un 1 a la columna correspondiente a la categoría del valor y un 0 en las otras columnas. Por ejemplo, si un registro tiene el color "Verde", en ese registro la columna "Color_Verde" tendría un valor de 1, mientras que "Color_Rojo" y "Color_Azul" tendrían valores de 0.

```python
cols_cat = ['job', 'marital', 
       'education', 'default',
       'housing', 'loan', 'contact', 'month',
       'day_of_week', 'poutcome']
df[cols_cat]
cols_new_cat=pd.get_dummies(df[cols_cat],drop_first = False)
cols_new_cat.head()
```

## Entrenamiento, validación y muestreo

Hasta ahora hemos explorado nuestros datos y creado columnas a partir de los datos categóricos. Ahora es el momento de dividir nuestros datos. La idea detrás de dividir los datos es que puedas medir qué tan bien funcionaría tu modelo con datos no vistos. Dividimos en tres partes:

Muestras de entrenamiento: estas son muestras del conjunto de datos que se utilizan para entrenar el modelo. Puede ser el 70% de los datos.

Muestras de validación: estas son muestras que se utilizan para validar o tomar decisiones a partir del modelo. Puede ser el 15% de los datos.

Muestras de prueba: estas son muestras que se utilizan para medir la precisión o el rendimiento del modelo. Puede ser el 15% de los datos.

En este proyecto, dividiremos en 70% entrenamiento, 15% validación y 15% prueba.

Vamos a mezclar las muestras utilizando sample en caso de que haya algún orden (por ejemplo, todas las muestras positivas en la parte superior). Aquí n es el número de muestras. random_state se especifica para que el proyecto sea reproducible.

Mezclando los datos

df_data.sample(n=len(df_data), random_state=42): Esta línea selecciona aleatoriamente todas las filas del DataFrame df_data. Aquí n=len(df_data) asegura que se seleccionen todas las filas, y random_state=42 asegura que la aleatorización sea reproducible, es decir, que siempre se obtenga el mismo resultado al ejecutar el código.

df_data.reset_index(drop=True): Esta línea restablece los índices del DataFrame después de que se han mezclado las filas. drop=True significa que el índice anterior se descarta y no se agrega como una columna al DataFrame.

```python
df_data = df_data.sample(n = len(df_data), random_state = 42)
df_data = df_data.reset_index(drop = True)
```

El número 42 en random_state=42 no tiene un significado especial en términos de funcionalidad del código. Es simplemente un valor elegido arbitrariamente para inicializar el generador de números aleatorios.

Tomemos ahora el 30% para validación y muestreo

```python
df_valid_test=df_data.sample(frac=0.30,random_state=42)
print('Tamnaño con respecto al dataset original: %.3f'%(len(df_valid_test)/len(df_data)))
```

Tomemos el 50% de este dataset para validación y el otro 50% para muestreo

```python
df_test = df_valid_test.sample(frac = 0.5, random_state = 42)
df_valid = df_valid_test.drop(df_test.index)
```

El 70% restante de los datos lo utilizaremos para entrenamiento

```python
df_train_all=df_data.drop(df_valid_test.index)
```

```python
# Revisando el balanceo de la variable Target 
print('Test(n = %d):%.3f'%(len(df_test),calc_porcentaje(df_test.OUTPUT_LABEL.values)))
print('Validación (n = %d):%.3f'%(len(df_valid),calc_porcentaje(df_valid.OUTPUT_LABEL.values)))
print('Entrenamiento (n = %d):%.3f'%(len(df_train_all), calc_porcentaje(df_train_all.OUTPUT_LABEL.values)))
```

Necesitamos equilibrar el conjunto de datos porque si usamos los datos de entrenamiento tal como están para el modelo predictivo, la precisión será muy alta, pero esto se debe a que no hemos capturado adecuadamente el resultado y que indica si una persona comprará un depósito a plazo o no. Hay más negativos que positivos, por lo que el modelo predictivo asignara negativos a muchas de las muestras.

Any missing values were filled with the mean value

```python
def fill_my_missing(df, df_mean, col2use):
    # This function fills the missing values

    # check the columns are present
    for c in col2use:
        assert c in df.columns, c + ' not in df'
        assert c in df_mean.col.values, c+ 'not in df_mean'
    
    # replace the mean 
    for c in col2use:
        mean_value = df_mean.loc[df_mean.col == c,'mean_val'].values[0]
        df[c] = df[c].fillna(mean_value)
    return df
```

## Model Selection 

This section allows us to test various  machine learning algorithm to see how our independent variables accurately predit our dependent y output variable. We will then select the best model based on performance on the validation set.
We will utilize the following functions to evaluate the performance of the model.

```python
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
def calc_specificity(y_actual, y_pred, thresh):
    # calculates specificity
    return sum((y_pred < thresh) & (y_actual == 0)) /sum(y_actual ==0)

def print_report(y_actual, y_pred, thresh):
    
    auc = roc_auc_score(y_actual, y_pred)
    accuracy = accuracy_score(y_actual, (y_pred > thresh))
    recall = recall_score(y_actual, (y_pred > thresh))
    precision = precision_score(y_actual, (y_pred > thresh))
    specificity = calc_specificity(y_actual, y_pred, thresh)
    f1 = 2 * (precision * recall) / (precision + recall)
   
    print('AUC:%.3f'%auc)
    print('accuracy:%.3f'%accuracy)
    print('recall:%.3f'%recall)
    print('precision:%.3f'%precision)
    print('f1:%.3f'%f1)
    print(' ')
    return auc, accuracy, recall, precision, specificity, f1
```

## Learning Curves

We can diagnose how our models are doing by plotting a learning curve. In this section, we will make use of the learning curve code from scikit-learn’s website with a small change of plotting the AUC instead of accuracy. Gradient Descent has the best AUC score (0.8) for the validation model and the learning curve for the model will be displayed below.

```python
import numpy as np
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and training learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 3-fold cross-validation,
          - integer, to specify the number of folds.
          - An object to be used as a cross-validation generator.
          - An iterable yielding train/test splits.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("AUC")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring = 'roc_auc')
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="b")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="b",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

The Stochastic Gradient Descent model with max_depth = 20, resulting in high variance

### Variance and Bias

In the case of random forest, we can see the model has high variance because the training and cross-validation scores show data points which are very spread out from one another. High variance would cause an algorithm to model the noise in the training set (overfitting).

Depending on the learning curve, there are a few strategies we can employ to improve the models

High Variance:
- Add more samples
- Add regularization
- Reduce number of features
- Decrease model complexity
- Add better features
- Change model architecture

High Bias:
- Add new features
- Increase model complexity
- Reduce regularization
- Change model architecture

## Hyperparameter tuning

A hyperparameter is a tool used in machine learning in order to estimate the model parameters (used in tuning a predictive modeling problem).  Hyperparameters are used in various machine learning algorithms. 

```python
# train a model for each max_depth in a list. Store the auc for the training and validation set

# max depths
max_depths = np.arange(2,20,2)

train_aucs = np.zeros(len(max_depths))
valid_aucs = np.zeros(len(max_depths))

for jj in range(len(max_depths)):
    max_depth = max_depths[jj]

    # fit model
    rf=RandomForestClassifier(n_estimators = 100, max_depth = max_depth, random_state = 42)
    rf.fit(X_train_tf, y_train)        
    # get predictions
    y_train_preds = rf.predict_proba(X_train_tf)[:,1]
    y_valid_preds = rf.predict_proba(X_valid_tf)[:,1]

    # calculate auc
    auc_train = roc_auc_score(y_train, y_train_preds)
    auc_valid = roc_auc_score(y_valid, y_valid_preds)

    # save aucs
    train_aucs[jj] = auc_train
    valid_aucs[jj] = auc_valid
```

n_estimators is a hyperparameter in the RandomForestClassifier that depending on the numbers of estimators entered the model can be overfitted, good compromise or underfitted. n_estimators is used for fine tuning the models in order to fit the training data. max_depth is also another hyperparameter; it controls the depth of the machine learning algorithm model.

```python
import matplotlib.pyplot as plt

plt.plot(max_depths, train_aucs,'o-',label = 'train')
plt.plot(max_depths, valid_aucs,'o-',label = 'valid')

plt.xlabel('max_depth')
plt.ylabel('AUC')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV

# number of trees
n_estimators = range(200,1000,200)
# maximum number of features to use at each split
max_features = ['auto','sqrt']
# maximum depth of the tree
max_depth = range(2,20,2)
# minimum number of samples to split a node
min_samples_split = range(2,10,2)
# criterion for evaluating a split
criterion = ['gini','entropy']

# random grid

random_grid = {'n_estimators':n_estimators,
              'max_features':max_features,
              'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'criterion':criterion}

print(random_grid)
```

```python
from sklearn.metrics import make_scorer, roc_auc_score
auc_scoring = make_scorer(roc_auc_score)
```
